# Connect to FHIR SQL Builder

```python
import intersystems_iris as iris
```

## Create the dbapi connection

```python
# create the dbapi connection

host = "iris-training"
port = 1972
namespace = "IRISAPP"
user = "SuperUser"
password = "SYS"

conn = iris.connect(
                    hostname=host, 
                    port=port, 
                    namespace=namespace, 
                    username=user, 
                    password=password
                   )

# create a cursor
cur = conn.cursor()
```

## Create the DriverLicense Table

```python
# execute a query
cur.execute("""create TABLE if not exists AB.DriverLicense (

    Code varchar(255) not null,
    Valide TINYINT not null,
    CreatedOn TIMESTAMP DEFAULT SYSDATE

)""")
```

## Insert New Data

```python
cur.execute("""delete AB.DriverLicense""")
```

```python
cur.execute("""INSERT INTO AB.DriverLicense (Code, Valide) VALUES ('S99916528', 1)""")
cur.execute("""INSERT into AB.DriverLicense (Code, Valide) VALUES ('S99950276', 0)""")
```

## Check new data

```python
cur.execute("select * from AB.DriverLicense")
cur.fetchall()
```

```python
# Solution propre et complète
import pandas as pd
from sqlalchemy import create_engine

# Créer l'engine SQLAlchemy
engine = create_engine('iris://' + user + ':' + password + '@' + host + ':' + str(port) + '/' + namespace)

sql = "select * from AB.DriverLicense"

# Méthode recommandée par pandas : utiliser SQLAlchemy avec context manager
with engine.connect() as connection:
    df = pd.read_sql(sql, connection)

print(df)
```
