```python
import pandas as pd

import matplotlib.pyplot as plt

# Réutilise une connexion déjà présente dans le notebook
sql_con = (
    globals().get("conn")
    or globals().get("connection")
    or globals().get("engine")
    or globals().get("iris_conn")
)

if sql_con is None:
    raise ValueError("Aucune connexion SQL trouvée (conn / connection / engine / iris_conn).")

# Essaie d'abord la classe correcte Ens.MessageHeader, puis la variante saisie Ens.MessageHeder
queries = [
    """
    SELECT SourceConfigName
    FROM Ens.MessageHeader
    WHERE SourceConfigName IS NOT NULL AND SourceConfigName <> ''
    """,
    """
    SELECT SourceConfigName
    FROM Ens.MessageHeder
    WHERE SourceConfigName IS NOT NULL AND SourceConfigName <> ''
    """
]

last_error = None
for query in queries:
    try:
        df_msg_header = pd.read_sql(query, sql_con)
        break
    except Exception as e:
        last_error = e
else:
    raise last_error

df_source_usage = (
    df_msg_header["SourceConfigName"]
    .value_counts()
    .rename_axis("SourceConfigName")
    .reset_index(name="nb_messages")
)

display(df_source_usage)

ax = df_source_usage.plot(
    kind="bar",
    x="SourceConfigName",
    y="nb_messages",
    figsize=(14, 5),
    legend=False,
    color="#4C78A8",
)

ax.set_title("Utilisation des SourceConfigName (ordre descendant)")
ax.set_xlabel("SourceConfigName")
ax.set_ylabel("Nombre de messages")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
